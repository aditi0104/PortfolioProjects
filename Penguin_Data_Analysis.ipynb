# Data Preproccessing

```python
df = df.dropna()
display(df)
```

```python
df['species'] = df.species.astype('category')
df['island'] = df.island.astype('category')
df['sex'] = df.sex.astype('category')
non_dummy_cols = ['island','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','sex','year']
dummy_cols = list(set(df.columns) - set(non_dummy_cols))
df_new = pd.get_dummies(df, columns=dummy_cols)
df_new['species_Adelie'].loc[df_new['species_Adelie'] == 1] = 0
df_new['species_Chinstrap'].loc[df_new['species_Chinstrap'] == 1] = 1
df_new['species_Gentoo'].loc[df_new['species_Gentoo'] == 1] = 2
df_new['species'] = df_new['species_Adelie'] + df_new['species_Chinstrap'] + df_new['species_Gentoo']
df_new = df_new.drop(columns=['species_Adelie','species_Chinstrap','species_Gentoo'])
df_new.head(10)
```

```python
df_new.head(10)
non_dummy_cols = ['species','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','sex','year']
dummy_cols = list(set(df.columns) - set(non_dummy_cols))
df_new = pd.get_dummies(df_new, columns=dummy_cols)
df_new['island_Biscoe'].loc[df_new['island_Biscoe'] == 1] = 0
df_new['island_Dream'].loc[df_new['island_Dream'] == 1] = 1
df_new['island_Torgersen'].loc[df_new['island_Torgersen'] == 1] = 2
df_new['island'] = df_new['island_Biscoe'] + df_new['island_Dream'] + df_new['island_Torgersen']
df_new = df_new.drop(columns=['island_Biscoe','island_Dream','island_Torgersen'])
df_new.head(10)
```

```python
df_new.head(10)
non_dummy_cols = ['species','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','island','year']
dummy_cols = list(set(df.columns) - set(non_dummy_cols))
df_new = pd.get_dummies(df_new, columns=dummy_cols)
df_new['sex_female'].loc[df_new['sex_female'] == 1] = 0
df_new['sex_male'].loc[df_new['sex_male'] == 1] = 1
df_new['sex'] = df_new['sex_female'] + df_new['sex_male']
df_new = df_new.drop(columns=['sex_female','sex_male'])
df_new.head(10)
```

```python
df_new.head(10)
non_dummy_cols = ['species','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','island','sex']
dummy_cols = list(set(df.columns) - set(non_dummy_cols))
df_new = pd.get_dummies(df_new, columns=dummy_cols)
df_new['year_2007'].loc[df_new['year_2007'] == 1] = 0
df_new['year_2008'].loc[df_new['year_2008'] == 1] = 1
df_new['year_2009'].loc[df_new['year_2009'] == 1] = 2
df_new['year'] = df_new['year_2007'] + df_new['year_2008'] + df_new['year_2009']
df_new = df_new.drop(columns=['year_2007','year_2008','year_2009'])
df_new.head(10)
```

```python
df_new['species'] = df_new.species.astype('category')
df_new['island'] = df_new.island.astype('category')
df_new['sex'] = df_new.sex.astype('category')
df_new['year'] = df_new.year.astype('category')
type(df_new['species'])
type(df_new['island'])
type(df_new['sex'])
type(df_new['year'])
df_new.dtypes
```
